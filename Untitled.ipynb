```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

class Object3D(object):
    def __init__(self, filename, target_class, train):
        self.filename = filename
        self.target_class = target_class
        self.train = train
        self.parse_file()
        self.total_surface = self.calc_total_surface()
        self.triangle_indices = np.arange(self.n_triangles)
        self.triangle_surfaces = [triangle.surface for triangle in self.triangles]
        
    def parse_file(self):
        with open(self.filename, 'r') as f:
            off_line = f.readline()
            if off_line[:3] != 'OFF':
                raise Exception('Not proper file')
            if len(off_line) > 4:
                self.n_vertices, self.n_triangles, _ = off_line[3:].split(' ')
            else:
                self.n_vertices, self.n_triangles, _ = f.readline().split(' ')
            self.n_vertices = int(self.n_vertices)
            self.n_triangles = int(self.n_triangles)
            self.vertices = []
            self.triangles = []
            for vertex in range(self.n_vertices):
                self.vertices.append(Vertex(f.readline()))
            for face in range(self.n_triangles):
                self.triangles.append(Triangle(f.readline(), self.vertices))

    def calc_total_surface(self):
        return sum([triangle.surface for triangle in self.triangles])
    
    def sample_points(self, n):
        sample_indices = np.random.choice(self.triangle_indices, size=n, replace=True, p=self.triangle_surfaces / self.total_surface)
        return PointCloud([self.triangles[s].sample_point() for s in sample_indices])
    
class PointCloud(object):
    def __init__(self, vertices):
        self.vertices = vertices
        
    def normalize(self, mean=None, std=None):
        if mean is None:
            mean = np.mean([v.np_array for v in self.vertices], axis=0)
            std = np.std([v.np_array for v in self.vertices], axis=0)
        if mean is not None:
            self.vertices = [(v - mean) / std for v in self.vertices]
            
    def project_axes(self, drop_axis=2):
        return np.array([np.delete(v.np_array, drop_axis) for v in self.vertices])
    
    def plot_projections(self):
        plt.figure(1, figsize=(15, 20))
        plt.subplot(3, 1, 1)
        plt.scatter(self.project_axes(0)[:, 0], self.project_axes(0)[:, 1], alpha=0.6)
        plt.axis('scaled')
        plt.subplot(3, 1, 2)
        plt.scatter(self.project_axes(1)[:, 0], self.project_axes(1)[:, 1], alpha=0.6)
        plt.axis('scaled')
        plt.subplot(3, 1, 3)
        plt.scatter(self.project_axes(2)[:, 0], self.project_axes(2)[:, 1], alpha=0.6)
        plt.axis('scaled')
        plt.show()        
    
class Vertex(object):
    def __init__(self, line=None, np_array=None):
        if line is not None:
            x, y, z = line.rstrip().split(' ')
            self.np_array = np.array([float(x), float(y), float(z)])
        else:
            self.np_array = np_array
            
    def __add__(self, other):
        return Vertex(np_array=self.np_array + other.np_array)
            
    def __sub__(self, other):
        if isinstance(other, Vertex):
            return Vertex(np_array=self.np_array - other.np_array)
        else:
            return Vertex(np_array=self.np_array - other)
    
    def __rmul__(self, other):
        return Vertex(np_array=self.np_array * other)
    
    def __truediv__(self, other):
        return Vertex(np_array=self.np_array / other)
        
class Triangle(object):
    def __init__(self, line, vertices):
        number, index1, index2, index3 = line.split(' ')
        if int(number) != 3:
            raise Exception('Non triangle encountered')
        index1, index2, index3 = int(index1), int(index2), int(index3)
        self.vertices = [vertices[index1], vertices[index2], vertices[index3]]
        self.surface = self.calc_surface()
        
    def calc_surface(self):
        return np.linalg.norm(np.cross(self.vertices[0].np_array - self.vertices[1].np_array, 
                              self.vertices[0].np_array - self.vertices[2].np_array)) / 2
    
    def sample_point(self):
        r1, r2 = np.random.uniform(size=2)
        return (1 - np.sqrt(r1)) * self.vertices[0] + np.sqrt(r1) * (1 - r2) * self.vertices[1] + np.sqrt(r1) * r2 * self.vertices[2]

def read_data(folder):
    categories = os.listdir(folder)
    train_objects = []
    test_objects = []
    counter = 0
    for category in categories[1:11]:
        counter += 1
        print(category, 'train')
        for file in os.listdir(os.path.join(folder, category, 'train')):
            train_objects.append(Object3D(os.path.join(folder, category, 'train', file), category, True))
        print(category, 'test')
        for file in os.listdir(os.path.join(folder, category, 'test')):
            test_objects.append(Object3D(os.path.join(folder, category, 'test', file), category, False))
    return train_objects, test_objects, categories
    
train_objects, test_objects, categories = read_data('data')
#read_data('data')
```

```python
## DeepSets network https://arxiv.org/pdf/1703.06114.pdf

class PointCloudNetwork(object):
    def __init__(self, n_points, n_classes, perm_layer_sizes, full_layers_sizes):
        
        def batch_vm(v, m):
            shape = tf.shape(v)
            rank = shape.get_shape()[0].value
            v = tf.expand_dims(v, rank)
            vm = tf.multiply(v, m)
            return tf.reduce_sum(vm, rank-1)
        
        self.n_points = n_points
        self.n_classes = n_classes
        self.perm_layer_sizes = perm_layer_sizes
        self.full_layer_sizes = full_layers_sizes
        
        self.input = tf.placeholder(dtype=tf.float32, shape=[None, n_points, 3])
        self.output = tf.placeholder(dtype=tf.int32, shape=[None])
        
        print('input', self.input.get_shape())
        print('output', self.output.get_shape())
        
        self.perm_W = []
        self.perm_b = []
        self.perm_layers = []
        
        last_input_size = 3
        last_input = self.input
        for perm_layer_size in self.perm_layer_sizes:
            self.perm_W.append(tf.Variable(tf.random_normal([last_input_size, perm_layer_size], stddev=0.35)))
            self.perm_b.append(tf.Variable(tf.zeros([perm_layer_size])))
            self.perm_layers.append(tf.nn.tanh(batch_vm(last_input, self.perm_W[-1]) + self.perm_b[-1]))
            last_input_size = perm_layer_size
            last_input = self.perm_layers[-1]
            print('last_input perm_layer', last_input.get_shape())
            
        self.pool_layer = tf.reduce_mean(last_input, -2)
        print('pool_layer', self.pool_layer.get_shape())
        
        last_input = self.pool_layer
        
        self.full_W = []
        self.full_b = []
        self.full_layers = []
        
        for full_layer_size in self.full_layer_sizes:
            self.full_W.append(tf.Variable(tf.random_normal([last_input_size, full_layer_size], stddev=0.35)))
            self.full_b.append(tf.Variable(tf.zeros([full_layer_size])))
            self.full_layers.append(tf.nn.tanh(tf.matmul(last_input, self.full_W[-1]) + self.full_b[-1]))
            last_input_size = full_layer_size
            last_input = self.full_layers[-1]
            
        self.out_W = tf.Variable(tf.random_normal([last_input_size, n_classes], stddev=0.35))
        self.out_b = tf.Variable(tf.zeros([n_classes]))
        self.output_logits = tf.matmul(last_input, self.out_W) + self.out_b
        self.output_layer = tf.nn.softmax_cross_entropy_with_logits(logits=self.output_logits, labels=self.output)
        
        self.loss = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=self.output_logits, labels=self.output)
        self.mean_loss = tf.reduce_mean(self.loss, -1)
        
        self.train = tf.train.AdamOptimizer().minimize(self.mean_loss)
        
        self.sess = tf.InteractiveSession()
        self.sess.run(tf.global_variables_initializer())
        
pcn = PointCloudNetwork(500, 10, [32, 24, 16], [16, 16])
```

```python
def count_number_trainable_params():
    '''
    Counts the number of trainable variables.
    '''
    tot_nb_params = 0
    for trainable_variable in tf.trainable_variables():
        shape = trainable_variable.get_shape() # e.g [D,F] or [W,H,C]
        current_nb_params = get_nb_params_shape(shape)
        tot_nb_params = tot_nb_params + current_nb_params
    return tot_nb_params

def get_nb_params_shape(shape):
    '''
    Computes the total number of params for a given shap.
    Works for any number of shapes etc [D,F] or [W,H,C] computes D*F and W*H*C.
    '''
    nb_params = 1
    for dim in shape:
        nb_params = nb_params*int(dim)
    return nb_params 

count_number_trainable_params()
```

```python
def accuracy(pcn, feed_dict, y_test):
    predictions = pcn.sess.run(pcn.output_logits, feed_dict=feed_dict)
    exp_predictions = np.exp(predictions)
    total_sum = np.sum(exp_predictions, axis=1)
    new_predictions = exp_predictions / total_sum.reshape(-1, 1)
    category_predict = np.argmax(new_predictions, axis=1)
    return np.sum(y_test == category_predict) / len(y_test)

accuracy(pcn, feed_dict_test, y_test)
```
